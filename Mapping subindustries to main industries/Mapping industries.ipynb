```python
import pandas as pd
```

```python
df = pd.read_csv('industry_list.csv')
```

```python
df
```

```python
df['Code'] = df['Industry'].map(lambda x: x.split(" ")[0].replace('[',' ').replace(']','').replace(' ',""))
df
```

```python
def make_ranges(row):
    try:
        if row == '00':
            return "0000"
        else:
            list_number = row.split("-")
            x = int(list_number[0])*100
            y = int(list_number[1])*100 +100
            return list(range(x,y))
    except IndexError:
        return "error"
```

```python
df['range'] = df['Code'].map(make_ranges)
df
```

```python
range_dict = {}

def make_dict(row):
    try:

        if row.range == "0000":
            range_dict[0] = row.Industry
        else:
            for x in row.range:
                range_dict[x] = row.Industry
    except ValueError:
        return "Error"
```

```python
df.apply(make_dict,axis=1)
range_dict.get(9106)
```

```python
df_sub = pd.read_csv('sub_industry.csv')
```

```python
df_sub.head()
```

```python
df_sub['Code'] = df_sub['sub_industry'].map(lambda x: x.split()[0].replace('[',"").replace(']',""))
```

```python
final_sub = df_sub[df_sub['Code'].str.len() > 3]
```

```python
def match_industry(row):
    industry = range_dict.get(int(row))
    return industry
```

```python
final_sub['Industry'] = final_sub['Code'].map(match_industry) 
```
