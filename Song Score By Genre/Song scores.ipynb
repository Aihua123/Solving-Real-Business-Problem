```python
import pandas as pd
df = pd.read_csv('class.csv')
df.head()
```

```python
df.info()
```

```python
result = {}
for index, row in df.iterrows():
    file_name = row.FILE_NAME
    temp = df.iloc[index][df.columns[4:]].to_dict()
    temp = {k:v for k,v in temp.items() if v> 0}
    result[file_name] = temp
```
